## Matplotlib Example Workflow

```python
# STEP 0: Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# STEP 1: Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# STEP 2: Setup Plot
fig, ax = plt.subplots(figsize = (5, 5)) #(Width, Height)

# STEP 3: Plot the data
ax.plot(x, y)

# STEP 4: Customize Plot
ax.set(title = "Sample Plot",
      xlabel = "X-Axis",
      ylabel = "Y-Axis")

# STEP 4: Save & Show
fig.savefig("Images/Sample_Plot.png")
```

### Bar Plot (Vertical & Horizontal)

```python
# Make a plot from Dictionary & Create Bar Plot
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "My Nut Butter Store",
      xlabel = "Butter",
      ylabel = "Prices");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
# For horizontal bar, we gotta convert it into list type
```

### Histogram Plot

```python
# Plot the data & Create Histogram Plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### SubPlot

#### Method 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

#### Method 2

```python
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

# Plot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## PyPlot V/S Matlplotlib OO Method

* When plotting something quickly, okay to use PyPlot Method
* When plotting something more advanced, use OO Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# PyPlot Method
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO Method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);

ax.set_xlim([45, 100])
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot Data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize Plot
ax.set(title = "Heart Disease & Cholestrol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a Legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal Line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

## NOW WE'RE GONNA MAKE THIS WHOLE THING

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2,
                               ncols = 1, 
                               figsize = (10, 10))

# ------------------ first plot ------------------

# Plot Data
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize Plot
ax1.set(title = "Heart Disease & Cholestrol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a Legend
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal Line
ax1.axhline(over_50["chol"].mean(),
           linestyle='--');

# ------------------ second plot ------------------

# Plot Data
scatter = ax2.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"])

# Customize Plot
ax2.set(title = "Heart Disease & Max Heart Levels",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# Add a Legend
ax2.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal Line
ax2.axhline(over_50["thalach"].mean(),
           linestyle='--');

# Add Title to the whole thing
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
