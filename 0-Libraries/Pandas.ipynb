### Create data

```python
# Column from Series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New Column called series
car_sales["Seats"] = seats_column
car_sales
```

#### You can enter less values if you create data from Series, it won't be a problem

```python
car_sales["Seats"].fillna(5, inplace = True)
car_sales
```

```python
# Column from Python list
fuel_eco = [7.5, 6.3, 3.5, 2.8, 8.4, 5.6, 6.8, 3.5, 2.8, 8.4]
car_sales["Fuel/100KM"] = fuel_eco
car_sales
```

##### You can't enter less values if you create data from Python List, it will show an error

```python
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel/100KM"]
car_sales
```

```python
car_sales.drop("Total Fuel Used", axis = 1, inplace = True)
```

```python
car_sales
```

### Shuffle Data

```python
car_sales.sample(frac=1)
# frac is how much part of data you wanna shuffle
# For eg if we wrote frac = 0.5, ir would have shuffled 50% of the data & shown that.
```

### Playing with a specific column (Here change unit from KM to miles)

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6)
car_sales
```
