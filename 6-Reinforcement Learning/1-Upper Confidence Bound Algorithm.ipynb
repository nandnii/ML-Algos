# UPPER CONFIDENCE BOUND ALGORITHM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
dataset
```

## Implementing UCB

```python
import math

N = 10000
d = 10
ads_selected = []
no_of_selections = [0] * d # This is  Ni(n)
sum_of_rewards = [0] * d # This is Ri(n)
total_reward = 0

for n in range(0, N):
    ad = 0
    max_upper_bound = 0
    for i in range (0, d):
        if no_of_selections[i] > 0: # Doing this to check bcz denominator of ri(n) should be > 0 
            average_reward = sum_of_rewards[i] / no_of_selections[i] # This is ri(n)
            delta_i = math.sqrt(3/2 * math.log(n + 1) / no_of_selections[i])
            upper_bound = average_reward + delta_i
        
        else:
            upper_bound = 1e400
            
        if(upper_bound > max_upper_bound):
            max_upper_bound = upper_bound
            ad = i
            
    ads_selected.append(ad)
    no_of_selections[ad] = no_of_selections[ad] + 1
    reward = dataset.values[n, ad]
    sum_of_rewards[ad] = sum_of_rewards[ad] + reward
    total_reward = total_reward + reward
```

## Visualizing Results

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
